```python
# Task 1: Exploring and Visualizing a Simple Dataset
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Dataset
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# Inspecting the Dataset
```

```python
# shape of dataset
print("Shape of dataset:", df.shape)

# column names
print("Columns:", df.columns)

# first 5 rows
df.head()
```

```python
#Summary Statistics
```

```python
# dataset info
df.info()

# statistical summary
df.describe()
```

```python
# Data Visualization

# Scatter Plot

#This scatter plot shows the relationship between sepal length and petal length, with different colors representing different species.
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
### Histogram

#Histograms show the distribution of values for each feature in the dataset.
```

```python
df.hist(figsize=(8,6))
plt.show()
```

```python
### Box Plot

#Box plots help identify outliers and show the spread of the data.
```

```python
sns.boxplot(data=df)
plt.title("Box Plot of Features")
plt.show()
```
